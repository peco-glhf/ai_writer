```python
import time  # â†è¿½åŠ 
start_time = time.time()  # â±ï¸ é–‹å§‹æ™‚åˆ»ã®è¨˜éŒ²
from google.colab import drive
drive.mount('/content/drive')

# å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
!apt install -y ffmpeg
!pip install noisereduce librosa soundfile pydub

# å„ç¨®ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import os
import datetime
import math
from google.colab import files
import noisereduce as nr
import librosa
import soundfile as sf
from pydub import AudioSegment, effects

# ãƒ™ãƒ¼ã‚¹ãƒ•ã‚©ãƒ«ãƒ€ã‚’æŒ‡å®šï¼ˆãƒ—ãƒ­ã‚¸ã‚§ã‚¯ãƒˆå˜ä½ã®ãƒ«ãƒ¼ãƒˆï¼‰
base_folder = '/content/drive/MyDrive/CLWORK/BL/éŸ³å£°å‡¦ç†çµæœ/éŸ³å£°å‡¦ç†/LG'
# ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰å‡¦ç†
uploaded = files.upload()
input_filename = list(uploaded.keys())[0]
file_basename = os.path.splitext(input_filename)[0]  # æ‹¡å¼µå­ãªã—ãƒ•ã‚¡ã‚¤ãƒ«å

# 10åˆ†ã‚»ã‚°ãƒ¡ãƒ³ãƒˆè¨­å®šï¼ˆç§’å˜ä½ï¼‰
segment_duration_sec = 10 * 60  # 10åˆ† = 600ç§’

# ã‚¿ã‚¤ãƒ ã‚¹ã‚¿ãƒ³ãƒ—ä»˜ãã®ä¸€æ„ãƒ•ã‚©ãƒ«ãƒ€ã‚’ä½œæˆ
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
drive_folder = f'{base_folder}/{file_basename}_{timestamp}'
os.makedirs(drive_folder, exist_ok=True)
os.makedirs(f'{drive_folder}/segments', exist_ok=True)  # ã‚»ã‚°ãƒ¡ãƒ³ãƒˆä¿å­˜ç”¨ãƒ•ã‚©ãƒ«ãƒ€

print(f"å‡¦ç†çµæœã¯ä»¥ä¸‹ã®ãƒ•ã‚©ãƒ«ãƒ€ã«ä¿å­˜ã•ã‚Œã¾ã™: {drive_folder}")

# ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ä¸€æ™‚ä¿å­˜
drive_audio_path = f'{drive_folder}/{input_filename}'
with open(input_filename, 'rb') as f_in:
    with open(drive_audio_path, 'wb') as f_out:
        f_out.write(f_in.read())


# éŸ³å£°ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ç›´æ¥èª­ã¿è¾¼ã¿ï¼ˆå‡¦ç†ãªã—ï¼‰
audio = AudioSegment.from_file(drive_audio_path)


# 10åˆ†ã”ã¨ã«ã‚»ã‚°ãƒ¡ãƒ³ãƒˆã«åˆ†å‰²ã—ã¦ä¿å­˜
total_duration_sec = len(audio) / 1000  # ãƒŸãƒªç§’ã‹ã‚‰ç§’ã«å¤‰æ›
num_segments = math.ceil(total_duration_sec / segment_duration_sec)


print(f"éŸ³å£°ã®ç·æ™‚é–“: {total_duration_sec:.2f}ç§’ ({total_duration_sec/60:.2f}åˆ†)")
print(f"åˆ†å‰²æ•°: {num_segments}å€‹ (å„{segment_duration_sec/60:.1f}åˆ†)")

segment_files = []

for i in range(num_segments):
    # ã‚»ã‚°ãƒ¡ãƒ³ãƒˆã®é–‹å§‹æ™‚é–“ã¨çµ‚äº†æ™‚é–“ï¼ˆãƒŸãƒªç§’ï¼‰
    start_ms = i * segment_duration_sec * 1000
    end_ms = min((i + 1) * segment_duration_sec * 1000, len(audio))

    # æ™‚é–“è¡¨è¨˜ã®ç”Ÿæˆï¼ˆä¾‹: 00-10min, 10-20minï¼‰
    start_min = int(start_ms / 1000 / 60)
    end_min = int(end_ms / 1000 / 60)
    time_label = f"{start_min:02d}-{end_min:02d}min"

    # ã‚»ã‚°ãƒ¡ãƒ³ãƒˆã®æŠ½å‡º
    segment = audio[start_ms:end_ms]

    # ã‚»ã‚°ãƒ¡ãƒ³ãƒˆãƒ•ã‚¡ã‚¤ãƒ«ã®ä¿å­˜
    segment_filename = f'{file_basename}_{time_label}.wav'
    segment_path = f'{drive_folder}/segments/{segment_filename}'
    segment.export(segment_path, format='wav')
    segment_files.append(segment_path)

    print(f"ã‚»ã‚°ãƒ¡ãƒ³ãƒˆ {i+1}/{num_segments} ä¿å­˜: {segment_path}")


# å‡ºåŠ›ç¢ºèª
print("\nâœ… éŸ³å£°å‡¦ç†å®Œäº†")
print(f"ğŸ“ ã‚ªãƒªã‚¸ãƒŠãƒ«: {drive_audio_path}")
# print(f"ğŸ“ ãƒã‚¤ã‚ºé™¤å»æ¸ˆ: {drive_clean_path}")
# print(f"ğŸ“ æœ€çµ‚å‡ºåŠ›ï¼ˆãƒ•ãƒ«ï¼‰: {drive_final_path}")
print(f"ğŸ“ åˆ†å‰²ãƒ•ã‚¡ã‚¤ãƒ« ({len(segment_files)}å€‹):")
for file in segment_files:
    print(f"  - {os.path.basename(file)}")
end_time = time.time()  # â±ï¸ çµ‚äº†æ™‚åˆ»
elapsed_time = end_time - start_time
minutes, seconds = divmod(elapsed_time, 60)

print(f"\nâ³ å‡¦ç†ã«ã‹ã‹ã£ãŸæ™‚é–“: {int(minutes)}åˆ† {seconds:.2f}ç§’")
```
